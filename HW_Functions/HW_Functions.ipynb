## Задание 1
```
Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.
```

````python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def main():
    while True:
        user_input_1 = input('Введите команду: ')
        if user_input_1 == 'p':
            # узнать владельца по номеру документа 
            print_document_owner()
        elif user_input_1 == 's':
            # узнать полку по номеру документа
            document_number = input('Введите номер документа: ')
            directory = find_directory(document_number)
            if directory is None:
                print('Документ не найден в базе')
            else:
                print ('Документ хранится на полке: ', directory)
        elif user_input_1 == 'l':
            # узнать полную информацию по документам
            show_all_documents()
        elif user_input_1 == 'ads':
            # добавить полку
            directory = input('Введите номер полки: ')
            was_directory_added = add_directory(directory)
            all_directories = ', '.join(directories.keys())
            if was_directory_added is True: 
                print(f'Полка добавлена. Текущий перечень полок: {all_directories}')
            else:
                print(f'Такая полка уже существует. Текущий перечень полок: {all_directories}')
        elif user_input_1 == 'ds':
            # удалить полку
            directory = input('Введите номер полки: ')
            delete_directory(directory)          
        elif user_input_1 == 'q':
            # прекратить работу
            print('Конец работы')
            break
        else:
            print('Неизвестная команда. Доступные команды: p, s, l, ads, ds, q')
            
def print_document_owner():
    document_number = input('Введите номер документа: ')
        
    owner = None
    for document in documents:
        if document_number == document['number']:
            owner = document['name']    
            break
    if owner is None:
        print('Документ не найден в базе')
    else:
        print('Владелец документа: ', owner)
    
def find_directory(document_number):
    """Возвращает имя полки, если документ найден,
    None, если не найден"""
    found_directory = None
    for directory, document_numbers in directories.items():
        if document_number in document_numbers:
            found_directory = directory
            break
    if found_directory is None:
        return None
    else:
        return found_directory
    
def show_all_documents():
    """Печатает все сведения о всех документах, включая номер полки"""
    for document in documents:
        directory = find_directory(document['number'])
        print(f"№: {document['number']}, тип: {document['type']}, владелец: {document['name']}, полка хранения: {directory}")       

def add_directory(directory):
    """
        Принимает номер полки, которую нужно добавить. 
        Возвращает True, если полка добавлена, False, если такая полка уже существует.
    """
    if directory in directories:  
        return False
    else:
        directories[directory] = []
        return True

def delete_directory(directory):
    """
        Принимает номер полки, которую нужно удалить. 
        Выводит результат, не возвращает значения.
    """
    if directory not in directories:
        print(f'Такой полки не существует. Текущий перечень полок: {", ".join(directories.keys())}')
    else:
        if len(directories[directory]) > 0:
            print(f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {", ".join(directories.keys())}')
        else:
            del(directories[directory])
            print(f'Полка удалена. Текущий перечень полок: {", ".join(directories.keys())}')
main()
````
